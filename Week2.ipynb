  - Low Abstraction amd Engineering facing: What are people doing?

  - Low abstraction: How to approach the implementation

## Agile and Product Management

As PM, take big idea and concepts, break it into smaller pieces enabling iteration and testing. Allows to test for value iteratively and quickly. Get value out quickly, test it.

## Design Thinking and the Product Manager

Question: **Who** is the user, and **how are they using the product**?. i.e. Personas and Problem Scenarios.

Finding the right problem, then iterating through to find the right solution is a tool of design thinking and parallels the pillars of design thinking:
- empathy
- creativity

### Personas, what do they:
- think
- feel
- see
- do

Make google "ad words" for each persona area, see if they lead to click through, i.e. step 1 of design. 

### Problem Scenarios
- What job(s) are you doing for the customer
- What existing need or behaviour are you fulfilling

What are the alternatives. Examine the alternatives to understand how the user relates to the alternative

Your value propositions, are they enough better (as opposed to just incremental) than the alternatives. takes you to a testable view

## Lean Startup and the PM

Avoid waste, keep team focussed on building something valuable for the customer. 

**Lean Startup = [Value propositions & assumptions] + [Customer discovery & Experiments]** (steps from Venture Design cycle)

About **testing motivation**. BJ Fogg's Behavioural model (see slides)
- y-axis = *Motivation*
- x-axis = *Ability* (sort of inverse of usability)
- Action curve

**Need to balance customer motivation with usability**. Focussing on usability but the customer is unmotivated = **failure**!

### Lean Startup is a way to test Motivation
- build
- measure
- learn

### Lean Startup - The Scientific Method
1. Idea
2. Hypothesis
3. Experimental Design
4. Experimentation
  - MUST have **metrics**
5. Pivot or Persevere
  - PIVOT = experiments disprove hypothesis, back to 1, OR
  - PERSEVERE = experiments prove hypothesis
  
The best experiments run in about a week. The best MVPs are really scrappy.

| MVP Archetype | Notes |
| --- |--- |
| Concierge    | Hand create the customer experience |
| Wizard of Oz | Fake the customer experience |
| Sales        | See if you can sell some |

### What is hard about Lean Startup?
- *I have devs now, I am going to lose them*
  - Earmark periods for testing ideas
  - Leverage devs in the process.
- *Prototyping feels more like real work, we are building something*
  - Not a good way to test motivation
- *Its weird running these experiments*.

## Managing (User) Habits: Trigger and Action

In reality, customers are not perfectly informed, not perfectly rational. A lot of PM has emotional basis. 

### The Hook Framework (see slides):
- **Trigger**
  - Internal
    - i browse Facebook cos I'm bored
  - External
    - I browse Facebook cos I get an email saying something is posted
- **Action**
  - Has a threshold, effort. i.e. phones make posting images to Facebook easy, lowered barrier to action.
  - Goal to reduce barrier to action
  - Barriers to action = 
    - Thinking that PMs and companies got lucky
    - Need to understand the psychology of the user
- **Variable reward**
  - Something the user gets, positive or negative. The unpredictible but tangible gratification from the **action**
  - In Facebook, might see a bunch of likes - or not
  - Creates habits around your product
- **Investment**
  - Actions that increase involvement preference, and load the next trigger
  - natural investment: the continued use of the product increases the users preference to use the product in the future
  - investment of data: If [say] you dont need to enter your friends address, they are already on facebook, etc. 

### Engagement happens by design. 
- Whats the internal trigger?
- Whats the external trigger?
- Whats the simplest behaviour they can do in anticipation of a reward
- Is the reward fulfilling, yet leaves them wanting more
- what does the user put back into the product in anticipation of a future reward

PM job is to **hypothesize what the rewards will be**. Then test them, amplify the positive rewards.

## User Story Mapping for PM

The Agile User Story is the centre of creating the right solution.

### User Story Format:
- **As as [persona]**
  - Who is this user
  - What makes them tick
  - Who is an example of such a person
- **I Want to [do something]**
- **so that I can [realise a reward]**
  - Why do they want to do this
  - What is the benefit/reward
  - How will we know if it is working
  
### Epic Stories
- Specific meaning in SaFE
- High level user stories
- Storyboard epics, i.e. how do users relate to the story

### Child stories
- More detailed stories under an Epic
- Features and User Stories in SaFE

## Prototyping for Product Managers

Everyone loves a prototype, feels like creating the real product but thats not the case.
- test usability
- assist with engineering
- find patterns between customers
- **CANNOT** test motivation

**Prototyping = Drive Better Interfaces**

## Prototyping cycle:
- User Stories 
- Prototypes (that map to the patterns)
  - **Always** create multiple concepts
- Usability testing
- Possibly back to re-frame User Stories

Look for user patterns, i.e. metaphors:
- Like eCommerce
- Like loyalty
- etc. 



---
## Usability Testing for Product Managers

### Failure mode = 
- **Question from PM:** Can we do more usability testing on this release?
- **Answer from Eng:** Tight schedule, wont have time, wont be able to make the changes

### Sequence the testing to avoid "no time at the end"
- **NOT** the same as validation testing 
- Three levels of testing:
  - Exploratory, then..
    - Pick a few of the apporaches you explored
  - Assessment, then..
    - Assess the approaches you explored, narrow dow to one
  - Validation
  
Create the usability test plan **BEFORE** going too deep with the prototypes, i.e. **TEST CASES**.

## Collaborating with Enterprise Customers

Big companies often try to manage small vendors like they are internal organisations. Need to create the right balance.

### Ideas:
1. **Anchor to Problems instead of Solutions**
  - Two anti-poles of **design failure**
    - Do exactly what the user asks
    - Assume you know best and ignore the user (big misconception about Steve Jobs on this one!!)
  - Solve the right problem. Question the discrete asks by understanding what motivated them
2. **Write fully narrated stories, really great User Stories** 
  - Stories = input for building stuff
3. **Work in prioritised small batches**
  - Get the right outcome before moving to the next step
4. **Test Appropriately and Often**
  - Order of testing:
    - Day -1: Is it valuable (before you even start building it, frontload the value)
    - Day 0: Does it work (before you roll it out, put real data in the system)
    - Day 30: Does it stick (are they using it, the way you expected them to)
    - Day 90: Did it matter (outcome vs forecast)
5. **Over-invest in Training**
  - Create guard rails around how products should be used
  - Needs finding
6. **Over-invest in Onboarding**
  - customer has a problem early on, sort it immediately
7. **Do retrospectives on everything**
8. **Design for Variation**
  - Especially with big customers
  - APIs are a good way to externalise variation
9. **If its NO, just say NO**
  - Offer alternatives, but be upfront with the NOs

## Shaping your Customer Funnel
- See slides
- Use AIDAOR as steps in the funnel

### Attention
- getting peoples attention
- Qualitative might be number of click throughs
  
### Interest
- Keeping peoples attention
- Doing well here means doing less, simplicity
- 5 second test = 
  - show some one a product or web page for 5 seconds 
  - ask a series of questions
  - can do this by email (lose interest or not quickly)
  
### Desire
- What is pushing the user to the next step, what underlying desires are we fulfilling

### Action
- customer takes action here
- minimise the amount of work for customer to buy product and get some reward
- examples:
  - Streamline contract process
  - Streamline online signon
  - etc.
- Quantitative metrics
  - No. customers going to page vs sigining on
  
### Onboarding
- What does user need to do to get reward from product
- Everything we need to do to make user a habitual user of product
  - Important with pay as you go SAAS, customer can drop anytime
- If onboarding is "stale" activity drops after a period of use
  - question the product
  
### Retention
- everything that happens once the user is a habitual user
- focus on your "engine of growth"
  - Sticky
  - Viral (organic), users trigger more users
  
## Different View:
- **GROWTH** = [Attention + Interest]
- **ENGAGEMENT** = [Interest + Desire + Action + Onboarding]
- **MONETIZATION** = [Onboarding + Retention]
