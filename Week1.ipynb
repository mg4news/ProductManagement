## Two approaches: Scale Friendly VS Innovation Friendly

Scale Friendly (legacy = waterfall) =
- start with big plan that details everything that needs to happen
- budget, cost optimise, solve distribution
- hold focus groups
- then.. launch

Innovation Friendly (desirability driven = agile) =
- identify customer desires, triggers
- test prototypes, get metrics
- add features based on observations

## Why is a Product Not A Product

Because its a **SOLUTION** (product + other stuff like services, etc). It provides an experience to the customer, you need to understand the end to end. Folding as much as possible into the product helps with feasibility and viability. 

Waterfall is not a good approach, **agile** is much more suited.

**Desirability = finding the right problem**

## Product Manager vs Product Owner

Agile has:
- Scrum master (agile coach) => maps to => traditional role of project manager
- Product Owner => maps to => product manager. (sometimes)
- Development team (= developers, testers, operations, etc)


| Area | Product Manager | Product Owner |
| --- |--- | --- |
| Focus | Business/Market (external)facing | Dev/Engineering (internal) facing |
| Create? | Focus, direction, road map | Story backlog |
| Time horizon | Program increment | iteration/sprint |
| Review | Features | User stories |

*Can* do both jobs, traditionally there is a split of effort. Can increase workload (~40%). Result in PM not spending enough time in market, product starts to suffer. 

### How do you know if PO is performing?:
- Are there are lot of surprises when the product is show to stakeholders
- Are there a lot of questions from dev like "why are we building this?"
- How robust is the design process (i.e. prototypes before code, etc)

### When should the Product Manager be the Product Owner?:
- How is the PO role performing, if poorly may need to step in
- What are the options, is someone else available
- Where id the business going to get the most value from you
- Can you work in small batches to see if it will work

### Good approaches:
- pair one or more POs with a PM, as a fixed team
- Immature market, really tight integration between PO and PM. Once market matures and procust starts scaling, roles should split. 
- Maturity of engineering can affect the decision, immaturity requires more PO oversight

## You and your Consulting and/or Support Team

We have our solution, i.e. product + services, 3rd party products AND consulting and support. Need to fold learnings from consulting/support back into core product. 

Support = How do we enhance the customer experience

Get to know your support team personally, follow up, regular contact and updates

There may be regular dips in customer journey, i.e. where customer is not happy. This is a good place to engage a a consulting product (e.g. Shopify, builds eCommerce sites for smal businesses), etc. 

Consulting = any paid engagement (internal and internal)

To close the loop, and to avoid becoming captive to consulting service, any learningsneed to be rolled into the product. Say we are getting swamped with support calls. Need to improve or change the product to reduce call load or make it easier to get in contact with support, rather than simply relying on consulting service to handle the load..

PMs should take support calls. Provided perspective.

## You and your Finance and Legal teams

### Working with Finance 
Issues:
- cant get metrics to understand revenue drivers

Work with finance to ensure:
- you get the right metircs
- you resolve revenue recognition 

#### Accrual vs Cash
- Cash is simple revenue, i.e. what was paid
  - e.g. USD1200 paid in January
- Accrual is cash matched to the period in which the expense occured
  - e.g. USD100 for month accrued over the entire year
  
#### Managerial vs Financial
- Managerial
  - inside/internal
  - forecast and planning
  - internal views of revenue and profitability
  - **deeply involved**
- Financial
  - External, communicate with public, tax, etc.
  - **Limited involvement**
  
### Working with Legal
- Work with Legal to make sure your deal points are covered in the agreements
